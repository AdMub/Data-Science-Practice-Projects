K Means Clustering Project

In this project, we will utilize the KMeans Clustering algorithm to categorize universities into two groups: Private and Public.

It's important to note that while we have labels for this dataset, we will not use them for the KMeans clustering algorithm, as it is an unsupervised learning method.

Typically, when using the KMeans algorithm, you do not have labels. In this instance, we will use the labels to evaluate the algorithm's performance. However, this is not standard practice for KMeans, so the classification report and confusion matrix at the end of this project don't accurately reflect a real-world scenario.

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop("Private", axis=1))
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

Creation of new column (cluster) by converting (1) for a Private and (2) for Public school

```python
def change(private):
    if private == "Yes":
        return 1
    else:
        return 0
```

```python
df["Cluster"] = df["Private"].apply(change)
```

```python
df.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(df["Cluster"],kmeans.labels_))
print("\n")
print(confusion_matrix(df["Cluster"],kmeans.labels_))
```
